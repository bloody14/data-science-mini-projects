** Student Grade Analyzer**





```python
#a NumPy array of 50 randomly generated scores (0–100)
import numpy as np

scores=np.random.randint(0,101,50)
scores
```

```python
mean=np.mean(scores)
mean
```

```python
median=np.median(scores)
median
```

```python
standard_deviation=np.std(scores)
standard_deviation
```

```python
percentile=np.percentile(scores, 50)
percentile
```

```python

maximum=np.max(scores)
print("Highest score is",maximum)
minimum=np.min(scores)
print("lowest score is:",minimum)
```

```python
#Boolean masking
failed_students=scores<40
print("failed students are:",failed_students)
print(scores[failed_students])
```

```python
distiction=scores>85
print("Distinction scores :",distiction)
print(scores[distiction]) #Boolean indexing.
print("Number of students with distinction:",np.sum(distiction))
```

```python
#normalization
normalized_scores=(scores-minimum)/(maximum-minimum)
print("normalized scores :",normalized_scores)
```

```python
#reshape
scores_reshaped=scores.reshape(5,10)
print("reshaped scores :",scores_reshaped)
```

**BONUS:- CALCULATE WEIGHTAGE**

```python
theory_marks=np.random.randint(0,100,size=50)
practical_marks=np.random.randint(0,51,size=50)
print("theory marks :",theory_marks)
print("practical marks :",practical_marks)
```

```python
theory_weightage=0.4
practical_weightage=0.6
final_scores=(theory_weightage*theory_marks)+(practical_weightage*practical_marks)
print("final score :",final_scores)
```
